# Merging datasets

## All timestamps variant

### Select data

```python
eye_states.head()
```

```python
eye_states_to_merge = eye_states[['timestamp', 'pupil_diameter']]
eye_states_to_merge.head()
```

```python
eye_states_to_merge.shape
```

```python
imu.head()
```

```python
imu_to_merge = imu[['timestamp', 'roll', 'pitch', 'yaw']]
imu_to_merge.head()
```

```python
imu_to_merge.shape
```

```python
gaze.head()
```

```python
gaze_to_merge = gaze[['timestamp', 'azimuth', 'elevation']]
gaze_to_merge.head()
```

```python
gaze_to_merge.shape
```

### Merge eye_states with gaze

```python
eye_states_gaze = eye_states_to_merge.merge(gaze_to_merge, on='timestamp')
```

```python
eye_states_gaze
```

```python
eye_states_gaze.shape
```

```python
eye_states_gaze_to_corr = eye_states_gaze[['pupil_diameter', 'azimuth', 'elevation']]
```

```python
eye_states_gaze_to_corr.corr()
```

```python
dataplot_eye_states_gaze = sb.heatmap(eye_states_gaze_to_corr.corr(), cmap="YlGnBu", annot=True, vmin=-1, vmax=1)
```

### Merge eye_states with imu with gaze

```python
eye_states_imu_gaze = eye_states_to_merge.merge(imu_to_merge, on='timestamp') \
    .merge(gaze_to_merge, on='timestamp')
```

```python
eye_states_imu_gaze
```

```python
eye_states_imu_gaze.shape
```

```python
eye_states_imu_gaze_to_corr = eye_states_imu_gaze[['pupil_diameter', 'roll', 'pitch', 'yaw', 'azimuth', 'elevation']]
```

```python
eye_states_imu_gaze_to_corr.corr()
```

```python
dataplot_eye_states_imu_gaze = sb.heatmap(eye_states_imu_gaze_to_corr.corr(), cmap="YlGnBu", annot=True, vmin=-1, vmax=1)
```
